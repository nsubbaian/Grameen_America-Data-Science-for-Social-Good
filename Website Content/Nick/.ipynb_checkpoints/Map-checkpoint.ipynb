```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
df_zipcodes = pd.read_excel('Branch Zip Codes.xlsx')
df = pd.read_parquet('all_data_retention_df')
renameCols(df)
df_h = pd.DataFrame()

fig = go.FigureWidget(make_subplots(rows=2, cols=1, specs=[[{'type':'scattergeo'}],[{'type':'histogram'}]],
                                   subplot_titles = ('Branches', 'Histogram')))

fig.add_trace(go.Scattergeo(
        lon = df_zipcodes['Long'],
        lat = df_zipcodes['Lat'],
        text = df_zipcodes['Branch'],
        mode = 'markers',
        marker = dict(size = 8, color = 'lightseagreen')
        ), 1,1)

fig.add_trace(go.Histogram(),2,1)


fig.update_geos(
    scope="usa",
    showcountries=True, countrycolor="White",
)


dd = widgets.Dropdown(
    options=df.columns,
    value=df.columns[0],
    description='Data Column:',
    disabled=False,
)

# Update histogram callback
def update_hist(trace, points, state):
#     col = 'costofHousing_Monthly'
    branches = fig.data[0].text[points.point_inds]
    global df_h
    df_h = df[df['Branch'].isin(branches)]
    fig.data[1].x = df_h[dd.value]
    fig.update_layout(annotations = [{},{'text':', '.join(branches) + ': ' + dd.value}], showlegend = False)
#     print(type(fig.data[1].x))

def update_hist_col(change):
    global df_h
    if len(df_h) != 0:
        fig.data[1].x = df_h[dd.value]
        fig.update_layout(annotations = [{},{'text':', '.join(np.unique(df_h['Branch'])) + ': ' + dd.value}], showlegend = False)

    
    
dd.observe(update_hist_col)

fig.data[0].on_click(update_hist)
# print(fig)
widgets.VBox([fig, dd])
# fig
# fig.show()
```
